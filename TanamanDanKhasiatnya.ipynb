```python
import pandas as pd
from PIL import Image
```

```python
df=pd.read_excel('tanaman.xls')
```

```python
import qrcode
logo = Image.open('skanawa.png')
for index, row in df.iterrows():
    qr = qrcode.QRCode(
        version=12,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=2,
        border=8
    )
    data="NAMA : "+row['NAMA']+"\n"+"LATIN : "+row['LATIN']+"\n"+"MANFAAT : "+row['MANFAAT']
    qr.add_data(data)
    qr.make()
    img = qr.make_image(fill_color="black", back_color="#ffffff")
    pos = ((img.size[0] - logo.size[0]) // 2, (img.size[1] - logo.size[1]) // 2)
    img.paste(logo, pos)
    if index<9 :
        filename = 'qr0'+str(index+1)+'.png'
    else :
        filename = 'qr'+str(index+1)+'.png'
    img.save('result/'+filename)
```
